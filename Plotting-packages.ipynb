```python
import numpy as np
import matplotlib as npl
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

def myline():
    n = input("How many points to plot? ")
    p = list()
    h = list()

    print("\nEnter x values: ")
    for i in range(int(n)):
        pn = input()
        p.append(int(pn))
        x = np.array(p)

    print("\nEnter y values: ")
    for i in range(int(n)):
        hn = input()
        h.append(int(hn))
        y = np.array(h)

    print("Generating the graph!")
    def f(color):
        plt.plot(x,y, c = color)
        plt.xlabel('x-axis')
        plt.ylabel('y-axis')
        plt.title('Interactive Graph')

    colors=['red','green','blue','black','cyan','yellow','magenta']
    interact(f, color = colors)


    plt. show()


def mybar():
    n = input("How many bar points to plot? ")
    p = list()
    h = list()
    t = list()

    print("\nEnter x values: ")
    for i in range(int(n)):
        pn = input()
        p.append(int(pn))
        x = np.array(p)
        
    print("\nEnter y values: ")
    for i in range(int(n)):
        hn = input()
        h.append(int(hn))
        y = np.array(h)
        

        
    print("\nEnter the Label for bars : ")
    for i in range(int(n)):
        tn = input()
        t.append(tn)
        tl = (t)

    colors2 = ['red','y','g','b','c','m','C6','k','C0','C1','C2','C3','C3','C4','C5']
    plt.bar(x,y,tick_label = tl ,width = 0.4, color = colors2)
    
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.title('Interactive Bar Graph')

    plt.show
    
    
def myPie_chart():
    s = int(input("Enter the number of slices: "))
    l = list()
    a = list()
    
    print("\nEnter the Portion Size: ")
    for i in range(int(s)):
        sn = input()
        a.append(int(sn))
        f = np.array(a)
        
    print("\nEnter the Label for slices : ")
    for i in range(int(s)):
        ln = input()
        l.append(ln)#int(ln))
        k = (l)#np.arrays(l)
    colors2 = ['red','y','g','b','c','m','w','k','C0','C1','C2','C3','C3','C4','C5']
    plt.pie(f,labels = k, colors = colors2,startangle = 90, shadow=True,radius = 1.2, autopct = '%1.1f%%')
    plt.legend()#,explode=(0,0,0.1,0)
    plt.show()
    
    

print("Choose the Mode:\n1.Line strip\n2.Bar chart\n3.Pie chart\n")

m = int(input())
if m == 1:
    myline()
elif m == 2:
    mybar()
elif m == 3:
    myPie_chart()
else:
    print("Invalid Choice!")
```
